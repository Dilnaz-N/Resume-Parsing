```python
#convert()
#annotate()
#data()

#store()

import re
from tika import parser
import pickle 

def convert(path):
    parsed_tika=parser.from_file(path)
    d=re.sub('\\s+', ' ', parsed_tika["content"])
    d=re.sub('•',"",d)
    print(d)
    return d

def annotate(text_data,list_of_entities,name_of_entity):
    location_of_entities=[]
    for i in list_of_entities:
        re_location=[]
        matches=re.finditer(i,text_data)
        for match in matches:
            re_location.append(list(match.span()))
        if len(re_location)!=0:
            for j in range(len(re_location)):
                re_location[j].append(name_of_entity)
                re_location[j]=tuple(re_location[j])
        location_of_entities+=re_location
    return location_of_entities

def data(text_data,list_of_locations,prev_list=[]):
    da=[]
    final_list=[]
    ent=dict()
    ent['entities']=list_of_locations
    final_list.append(text_data)
    final_list.append(ent)
    da.append(tuple(final_list))
    return da+prev_list

def store(data):
    with open('aravind.pkl', 'wb') as f:
        pickle.dump(data, f)
        
```
